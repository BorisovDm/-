## Введение в метод конечных элементов

## Тезисы
__Подходы к дискретизации__ при построении численных методов:
1. Дискретизация в расчетной области
 * метод конечных разностей (МКР): численное решение задано в конечном числе точек множества расчетной области
 * метод конечных объемов (МКО): численное решение задано в конечном числе подмножеств расчетной области
 
2. Дискретизация в фукциональных пространтствах
 * проекционные методы: численное решение задано во всей расчетной области; численное решение представляет собой разложение по базису конечномерного подпространтсва функционального пространства, которому принадлежит аналитическое решение.

Метод конечных элементов (МКЭ) - подвид проекционных методов (проекционно-сеточный метод).

* слабая (вариационная) постановка задачи для ОДУ; ограничения, накладывающиеся на тестовые функции
* лемма Лакса-Милграма, гарантирующая единственность решения задачи в слабой постановке
* доказательство сходимости метода Галеркина; требование полноты системы функций
* метод конечных элементов: пространство кусочно-линейных функций, базис в пространстве
* алгоритм ассемблирования матриц жесткости, демпфирования, пр. Использование квадратурных формул, решение СЛАУ
* итоговые замечания по МКЭ: выгода в сравнении с МКР, МКО.
* __обзор ООП реализации в симуляторе электрофизиологии Chaste__
