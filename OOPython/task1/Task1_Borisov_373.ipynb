```python
class Person:
    """Class Person with
    
    fields[name, mobilePhoneNumber(list), workPhoneNumber(list), homePhoneNumber, email(list), address, skype]
    
    and methods[SetName(name), SetMobilePhoneNumber(number), SetWorkPhoneNumber(number), SetHomePhoneNumber(number),\
    DelMobilePhoneNumber(number), DelWorkPhoneNumber(number), DelHomePhoneNumber(number), SetEmail(email),\
    DelEmail(email), SetAddress(address), SetSkype(skype), GetAllInformation()]
    """

    def __init__(self):
        self.__name = ''
        self.__mobilePhoneNumber = []
        self.__workPhoneNumber = []
        self.__homePhoneNumber = ''
        self.__email = []
        self.__address = ''
        self.__skype = ''
        
    def SetName(self, name):
        self.__name = name
        
    def SetMobilePhoneNumber(self, number):
        self.__mobilePhoneNumber.append(number)
        
    def SetWorkPhoneNumber(self, number):
        self.__workPhoneNumber.append(number)
        
    def SetHomePhoneNumber(self, number):
        self.__homePhoneNumber = number
        
    def DelMobilePhoneNumber(self, number):
        if number in self.__mobilePhoneNumber:
            self.__mobilePhoneNumber.remove(number)
        
    def DelWorkPhoneNumber(self, number):
        if number in self.__workPhoneNumber:
            self.__workPhoneNumber.remove(number)
        
    def DelHomePhoneNumber(self, number):
        if number == self.__homePhoneNumber:
            self.__homePhoneNumber = ''
            
    def SetEmail(self, email):
        self.__email.append(email)
    
    def DelEmail(self, email):
        if email in self.__email:
            self.__email.remove(email)
    
    def SetAddress(self, address):
        self.__address = address
        
    def SetSkype(self, skype):
        self.__skype = skype
        
    def GetAllInformation(self):
        if self.__name == '':
            print('Name: no information')
        else:
            print('Name: ' + self.__name)

        if len(self.__mobilePhoneNumber) == 0:
            print('Mobile Phone Number: no information')
        else:
            print('Mobile Phone Number: ' + ', '.join(self.__mobilePhoneNumber))
        
        if len(self.__workPhoneNumber) == 0:
            print('Work Phone Number: no information')
        else:
            print('Work Phone Number: ' + ', '.join(self.__workPhoneNumber))
        
        if self.__homePhoneNumber == '':
            print('Home Phone Number: no information')
        else:
            print('Home Phone Number: ' + self.__homePhoneNumber)
        
        if len(self.__email) == 0:
            print('E-mail: no information')
        else:
            print('E-mail: ' + ', '.join(self.__email))
       
        if self.__address == '':
            print('Address: no information')
        else:
            print('Address: ' + self.__address)
                    
        if self.__skype == '':
            print('Skype: no information')
        else:
            print('Skype: ' + self.__skype)
```

```python
names = ['Peter Brandt', 'Paul Nicholson', 'Nick Jones', 'Jack Smith']
workPhoneNumbers = ['+79871236543', '405-67-89', '+7(8521)6790']
mobilePhoneNumbers = ['456-345', '123-8765', '0987-23', '24-24']
homePhoneNumbers = ['356', '9876', '009876', '3456']
addressList = ['614 Bernardo Street', '1411 Lonely Oak Drive', '2214 Brannon Avenue', '2395 Chapmans Lane']

persons = []

for name in names:
    persons.append(Person())
    
for person, name, mobilePhone, homePhone, address \
in zip(persons, names, mobilePhoneNumbers, homePhoneNumbers, addressList):
    person.SetName(name)
    person.SetEmail(name.replace(' ', '') + '@gmail.com')
    person.SetSkype(name.replace(' ', ''))
    person.SetMobilePhoneNumber(mobilePhone)
    person.SetHomePhoneNumber(homePhone)
    person.SetAddress(address)
    for number in workPhoneNumbers:
        person.SetWorkPhoneNumber(number)
    
for person in persons:
    person.GetAllInformation()
    print(' ')
```
